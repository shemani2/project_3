```python
import requests
from datetime import datetime
import pandas as pd
from config import api_key


# Function to fetch COVID data
def fetch_covid_data(url):
    try:
        response = requests.get(url)
        response.raise_for_status()  # Raise an exception for bad status codes
        return response.json()
    except requests.exceptions.RequestException as e:
        print(f"Error fetching data: {e}")
        return None

# Function to extract 'currentUsageCovid' data for March of specified years
def extract_current_usage_covid(data, years):
    current_usage_covid = []
    for county_data in data:
        county_name = county_data["county"]
        for date_entry in county_data["actualsTimeseries"]:
            date = datetime.strptime(date_entry.get("date"), '%Y-%m-%d')
            if date.year in years and date.month == 3:
                hospital_beds = date_entry.get("hospitalBeds", {})
                if hospital_beds and 'currentUsageCovid' in hospital_beds:
                    current_usage_covid.append({
                        "County": county_name,
                        "Date": date.strftime('%Y-%m-%d'),
                        "Hospitalizations": hospital_beds['currentUsageCovid']
                    })
    return current_usage_covid

# Function to create DataFrame
def create_dataframe(data):
    return pd.DataFrame(data)

# URL for COVID data
url = "https://api.covidactnow.org/v2/county/TX.timeseries.json"

# Years to extract data for
years = [2021, 2022, 2023]

# Fetch data for March of 2021
start_date_2021 = datetime(2021, 3, 1)
end_date_2021 = datetime(2021, 3, 31)
query_url_2021 = f"{url}?apiKey={api_key}&start={start_date_2021.strftime('%Y-%m-%d')}&end={end_date_2021.strftime('%Y-%m-%d')}"
covid_data_2021 = fetch_covid_data(query_url_2021)

# Fetch data for March of 2022
start_date_2022 = datetime(2022, 3, 1)
end_date_2022 = datetime(2022, 3, 31)
query_url_2022 = f"{url}?apiKey={api_key}&start={start_date_2022.strftime('%Y-%m-%d')}&end={end_date_2022.strftime('%Y-%m-%d')}"
covid_data_2022 = fetch_covid_data(query_url_2022)

# Fetch data for March of 2023
start_date_2023 = datetime(2023, 3, 1)
end_date_2023 = datetime(2023, 3, 31)
query_url_2023 = f"{url}?apiKey={api_key}&start={start_date_2023.strftime('%Y-%m-%d')}&end={end_date_2023.strftime('%Y-%m-%d')}"
covid_data_2023 = fetch_covid_data(query_url_2023)

# Extract 'currentUsageCovid' data for each year
current_usage_covid_2021 = extract_current_usage_covid(covid_data_2021, years)
current_usage_covid_2022 = extract_current_usage_covid(covid_data_2022, years)
current_usage_covid_2023 = extract_current_usage_covid(covid_data_2023, years)

# Create DataFrames
df_2021 = create_dataframe(current_usage_covid_2021)
df_2022 = create_dataframe(current_usage_covid_2022)
df_2023 = create_dataframe(current_usage_covid_2023)

# Print or use DataFrames as needed
print(df_2021)
print(df_2022)
print(df_2023)
```

```python
import plotly.graph_objs as go
import ipywidgets as widgets
from IPython.display import display
import requests
from datetime import datetime
import pandas as pd
from config import api_key

# Function to fetch COVID data
def fetch_covid_data(url):
    try:
        response = requests.get(url)
        response.raise_for_status()  # Raise an exception for bad status codes
        return response.json()
    except requests.exceptions.RequestException as e:
        print(f"Error fetching data: {e}")
        return None

# Function to extract 'currentUsageCovid' data for March of specified years
def extract_current_usage_covid(data, years):
    current_usage_covid = []
    for county_data in data:
        county_name = county_data["county"]
        for date_entry in county_data["actualsTimeseries"]:
            date = datetime.strptime(date_entry.get("date"), '%Y-%m-%d')
            if date.year in years and date.month == 3:
                hospital_beds = date_entry.get("hospitalBeds", {})
                if hospital_beds and 'currentUsageCovid' in hospital_beds:
                    current_usage_covid.append({
                        "County": county_name,
                        "Date": date.strftime('%Y-%m-%d'),
                        "Hospitalizations": hospital_beds['currentUsageCovid']
                    })
    return current_usage_covid

# Function to create DataFrame
def create_dataframe(data):
    return pd.DataFrame(data)

# URL for COVID data
url = "https://api.covidactnow.org/v2/county/TX.timeseries.json"

# Years to extract data for
years = [2021, 2022, 2023]

# Fetch data for March of 2021
start_date_2021 = datetime(2021, 3, 1)
end_date_2021 = datetime(2021, 3, 31)
query_url_2021 = f"{url}?apiKey={api_key}&start={start_date_2021.strftime('%Y-%m-%d')}&end={end_date_2021.strftime('%Y-%m-%d')}"
covid_data_2021 = fetch_covid_data(query_url_2021)

# Fetch data for March of 2022
start_date_2022 = datetime(2022, 3, 1)
end_date_2022 = datetime(2022, 3, 31)
query_url_2022 = f"{url}?apiKey={api_key}&start={start_date_2022.strftime('%Y-%m-%d')}&end={end_date_2022.strftime('%Y-%m-%d')}"
covid_data_2022 = fetch_covid_data(query_url_2022)

# Fetch data for March of 2023
start_date_2023 = datetime(2023, 3, 1)
end_date_2023 = datetime(2023, 3, 31)
query_url_2023 = f"{url}?apiKey={api_key}&start={start_date_2023.strftime('%Y-%m-%d')}&end={end_date_2023.strftime('%Y-%m-%d')}"
covid_data_2023 = fetch_covid_data(query_url_2023)

# Extract 'currentUsageCovid' data for each year
current_usage_covid_2021 = extract_current_usage_covid(covid_data_2021, years)
current_usage_covid_2022 = extract_current_usage_covid(covid_data_2022, years)
current_usage_covid_2023 = extract_current_usage_covid(covid_data_2023, years)

# Create DataFrames
df_2021 = create_dataframe(current_usage_covid_2021)
df_2022 = create_dataframe(current_usage_covid_2022)
df_2023 = create_dataframe(current_usage_covid_2023)

# Function to update the plot based on the selected year
def update_plot(year):
    if year == 2021:
        df = df_2021
    elif year == 2022:
        df = df_2022
    elif year == 2023:
        df = df_2023
    
    fig = go.Figure()
    fig.add_trace(go.Scatter(x=df['Date'], y=df['Hospitalizations'], mode='lines+markers', name='Total Hospitalizations'))
    fig.update_layout(title='Hospitalizations Over Time', xaxis_title='Date', yaxis_title='Total Hospitalizations')
    fig.show()


# Create a dropdown widget for selecting the year
year_dropdown = widgets.Dropdown(options=[2021, 2022, 2023], value=2021, description='Year:')

# Register the update_plot function with the dropdown widget
widgets.interactive(update_plot, year=year_dropdown)
```
